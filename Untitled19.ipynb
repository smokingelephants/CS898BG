```python
import tensorflow as tf
import time
```

```python
start = time.time()

x = tf.constant(5.0)
w0 = tf.Variable(1.)
w1 = tf.Variable(1.)
record_w = []
record_y = []
for iter in range(3000):
  #print('iteration: ', iter)
  with tf.GradientTape(persistent=True) as g:
    g.watch(w0)
    g.watch(w1)
    with tf.GradientTape(persistent=True) as gg:
      gg.watch(x)
      y = w0 * x * x + tf.sin(w1*x) + tf.sin(w0*x)
      t = (-60-y) * (-60-y)
    dt_dx = gg.gradient(t, x)  # dy_dx = 2 * x
  #d2y_dx2 = g.gradient(dy_dx, x)  # d2y_dx2 = 2
  d2t_dw0 = g.gradient(dt_dx, w0)  # d2y_dx2 = 2
  d2t_dw1 = g.gradient(dt_dx, w1)  # d2y_dx2 = 2
  #print(dt_dx)

  ##print(d2y_dx2)
  #print(d2t_dw)
  #print(y)
  lr = 3e-6
  w0 = w0 - lr*d2t_dw0
  w1 = w1 - lr*d2t_dw1
  #print(w)
  #record_w.append(w.numpy())
  record_y.append(y.numpy())
  #print()

end = time.time()

#print(record_w)
print(record_y)
print(end - start)
t0 = end - start
```
